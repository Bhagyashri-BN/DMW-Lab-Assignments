```python
dataset = [
    ['cake', 'pizza', 'popsicle', 'orange juice', 'Cola', 'marshmallow', 'pie'],
    ['cake', 'Ice cream', 'peanut butter', 'orange juice', 'Cola', 'Ice cream'],
    ['cake', 'pie', 'nuts', 'orange juice', 'marshmallow', 'pie', 'pizza'],
    ['espresso', 'pizza', 'popsicle', 'orange juice', 'Cola', 'marshmallow'],
    ['cake', 'peanut butter', 'orange juice', 'Cola', 'pie'],
    ['espresso','pizza', 'popsicle', 'orange juice', 'Cola', 'marshmallow', 'pizza'],
    ['cake', 'peanut butter', 'Ice cream', 'orange juice', 'Cola', 'pie'],
    ['cake', 'pie', 'nuts', 'orange juice', 'marshmallow', 'pizza'],
    ['espresso', 'pizza', 'popsicle', 'orange juice', 'Cola', 'marshmallow', 'pie'],
    ['cake', 'peanut butter', 'orange juice', 'Cola', 'pie'],
    ['nuts', 'pizza', 'pizza', 'orange juice', 'Ice cream', 'Cola', 'pie']
]
```

```python
dataset
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
!pip install mlxtend
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

#**Association Rule Mining**

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.7)
res
```

```python
res1 = res[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
res1
```

```python
res2 = res1[res1['confidence']>=1]
res2
```
